# Задача A. Делители - количество пар на отрезке (сумма)

Заданы два натуральных числа $m$ и $n\left(1 \leq m<n \leq 10^8\right)$.

Рассмотрим все делители числа $n$. Нас будут интересовать только те делители, которые лежат на отрезке $[m ; n]$.

Требуется посчитать количество таких пар целых чисел $(a, b)$, что выполняются условия:
- $a$ и $b$ - делители числа $n$;
- $m \leq a<b \leq n$;
- $\operatorname{gcd}(a, b)=1$ (числа $a$ и $b$ взаимно простые);
- $a+b \leq n$.

## Формат ввода
Вводятся два натуральных числа $m$ и $n$.

## Формат вывода
Выведите одно целое число - количество таких пар.

## Решение

Задача очень похожа на задачу из демоверсии. Поэтому пойдем похожим путем — сначала найдем все делители числа $n$

````python
from math import gcd

m, n = map(int, input().split())

divs = []
for i in range(1, int(n ** 0.5) + 1):
    if n % i == 0:
        divs.append(i)
        if i != n // i:
            divs.append(n // i)
````

Далее нам нужно взять только те делители, которые удовлетворяют условию $d\geqslant m$

````python
divs = sorted([d for d in divs if d >= m])
````

Так как цикл по $j$ идет в промежутке $[i+1,\ \text{len(divs)})$ мы выполним условие $a<b$

````python
cnt = 0

for i in range(len(divs)):
    a = divs[i]
    for j in range(i + 1, len(divs)):
        b = divs[j]
        if a + b > n:
            break
        if gcd(a, b) == 1:
            cnt += 1

print(cnt)
````

# Задача B. Кумулятивная частота цифр

Дан массив целых чисел $a_1, a_2, \ldots, a_N$, каждый элемент которого лежит в диапазоне от 0 до 9 включительно. Для каждого целого числа $x$ от 0 до 9 определим значение $f(x)$ как количество элементов массива, которые не превосходят $x$.

Например $f(5)$ — это количество элементов массива, которые имеют значения от 0 до 5 включительно. Требуется для каждого целого числа $x$ от 0 до 9 посчитать $f(x)$.

## Формат ввода
В первой строке задано одно целое число $N$ - количество элементов массива ($1 \leqslant N \leqslant 1000000$).

Во второй строке заданы $N$ целых чисел $a_i\left(0 \leqslant a_i \leqslant 9\right)$.

## Формат вывода
Выведите 10 целых чисел - для каждого $x$ от 0 до 9 включительно значение $f(x)$. Числа необходимо выводить в одной строке, разделяя пробелами. Первое число должно соответствовать значению $x=0$, второе - значению $x=1$, ..., десятое число - значению $x=9$.

## Решение

````python
n = int(input())
a = list(map(int, input().split()))
````

Задача несложная. Давайте сначала насчитаем сколько раз появляются числа в массиве

````python
cnt = [0] * 10
for v in a:
    cnt[v] += 1
````

По условию, $f(x)$ — это количество элементов, которые не превосходят $x$, то есть $f(x)=cnt[0]+cnt[1]+\ldots+cnt[x]$ (что-то типа префиксных сумм)

````python
s = 0
ans = []
for x in range(10):
    s += cnt[x]
    ans.append(s)

print(*ans)
````

# Задача C. Мощность экспериментальной установки

Эдгар проводит эксперимент в лаборатории "Атомной энергетики" НИУ ВШЭ. Он моделирует "уровень мощности" портативного ядерного реактора, предложенного "Ассоциацией малой энергетики".

В начале эксперимента мощность реактора равна 1.
При моделировании, каждой день, он может выбрать ровно одну из трёх операций над текущей мощностью:
- увеличить мощность на 2;
- увеличить мощность на 3;
- увеличить мощность в 2 раза.

Все операции выполняются мгновенно, но между любыми двумя операциями обязательно проходит ровно одни сутки.

По заданному числу $N$ требуется определить, за какое минимальное количество суток Эдгар сможет получить мощность ровно $N$, начиная с единицы.

Если получить мощность ровно $N$ невозможно, нужно вывести -1.

## Формат ввода
Во входном файле задано одно целое число $N\left(1 \leqslant N \leqslant 10^6\right)$ - требуемое значение мощности.

## Формат вывода
Выведите одно целое число - минимальное количество суток, необходимое для получения мощности $N$, или -1, если это невозможно.

## Решение

````python
n = int(input())
````

Здесь используем типовое динамическое программирование. `dp[x]` будет хранить минимальное количество суток, за которое можно получить мощность ровно `x`, начиная с 1

Если `x` недостижимо, то `dp[x]=inf`

````python
inf = 10**18
dp = [inf] * (n + 1)
dp[1] = 0
````

За один день мы можем сделать
- либо +2
- либо +3
- либо x2

Если мы уже умеем получать значение `i` за `dp[i]` дней, то за ещё один день (`dp[i] + 1`) можем получить:
- `i + 2`
- `i + 3`
- `2 * i`

Так как одно и то же значение мощности можно получить разными путями, и они могут требовать разное число суток, нам нужно брать минимумы:
```python
dp[i + 2] = min(dp[i + 2], dp[i] + 1)
dp[i + 3] = min(dp[i + 3], dp[i] + 1)
dp[2 * i] = min(dp[2 * i], dp[i] + 1)
```



````python
for i in range(1, n + 1):
    if dp[i] == inf:
        continue
    if i + 2 <= n:
        dp[i + 2] = min(dp[i + 2], dp[i] + 1)
    if i + 3 <= n:
        dp[i + 3] = min(dp[i + 3], dp[i] + 1)
    if 2 * i <= n:
        dp[2 * i] = min(dp[2 * i], dp[i] + 1)

print(-1 if dp[n] == inf else dp[n])
````

# Задача D. Робот и минимальное число монет

Робот-сборщик монет работает в прямоугольной лаборатории, разбитой на клетки размером $N \times M$. Некоторые клетки содержат монеты, а некоторые клетки являются непроходимыми стенами.

Робот начинает движение в левой верхней клетке таблицы $(1,1)$ и хочет добраться в правую нижнюю клетку $(N, M)$.

За один ход робот может переместиться только:
- на одну клетку вправо,
- или на одну клетку вниз.

Если клетка является непроходимой, робот не может на неё войти.
При проходе через каждую проходимую клетку (включая стартовую и конечную) робот собирает все монеты, лежащие в этой клетке.

Требуется спланировать маршрут так, чтобы суммарное число собранных монет было минимально возможным. Если добраться из $(1,1)$ в $(N, M)$ невозможно, необходимо вывести -1.

## Формат ввода
В первой строке заданы два целых числа $N$ и $M$ - число строк и столбцов таблицы ($1 \leq N, M \leq 1000$).
Далее следуют $N$ строк, в каждой из которых записано по $M$ целых чисел.
- Если значение в клетке равно -1 , то эта клетка является непроходимой (стеной).
- Иначе в клетке лежит некоторое количество монет - натуральное число от 1 до $10^9$.

## Формат вывода
Выведите одно целое число - минимальное количество монет, которое робот может собрать по пути из клетки $(1,1)$ в клетку $(N, M)$.

Если такого пути не существует, выведите -1.

## Решение

Пусть `dp[i][j]` — минимальное число монет, которое можно собрать, дойдя до клетки `(i,j)`. Переход возможен только сверху или слева.

Обозначим за `inf` стену

````python
n, m = map(int, input().split())
a = [list(map(int, input().split())) for _ in range(n)]

inf = 10**30
dp = [[inf] * m for _ in range(n)]

if a[0][0] != -1:
    dp[0][0] = a[0][0]
````

Теперь нужно заполнить первые столбец и строку. С ними все просто — значение в ячейке будет равно сумме предыдущей ячейке и стоимости текущей

````python
for i in range(1, n):
    if a[i][0] != -1 and dp[i-1][0] != inf:
        dp[i][0] = dp[i-1][0] + a[i][0]

for j in range(1, m):
    if a[0][j] != -1 and dp[0][j-1] != inf:
        dp[0][j] = dp[0][j-1] + a[0][j]
````

Осталось заполнить оставшуюся матрицу по принципу `dp[i][j] = min(dp[i-1][j], dp[i][j-1]) + + a[i][j]`

Единственная проблема - нужно аккуратно обработать случаи, когда клетка, которую мы пытаемся заполнить является стеной и когда клетка слева или сверху являются стенами

````python
for i in range(1, n):
    for j in range(1, m):
        if a[i][j] != -1:
            best = min(dp[i-1][j], dp[i][j-1])
            if best != inf:
                dp[i][j] = best + a[i][j]

print(-1 if dp[n-1][m-1] == inf else dp[n-1][m-1])
````

# Задача E. Плитка

Студент факультета дизайна НИУ ВШЭ оформляет большую прямоугольную стену размерами $N \times M$ сантиметров. Он хочет использовать квадратную плитку одного размера, чтобы стена выглядела аккуратно.

Плитки должны:
- быть квадратами со стороной $a$ сантиметров (целое число);
- полностью лежать внутри стены;
- не перекрываться;
- быть ориентированы параллельно сторонам стены.

Часть стены может остаться не покрытой плиткой. Студенту нужно уложить как минимум $K$ плиток. Требуется подобрать такой максимальный возможный размер стороны $a$, при котором это возможно.

Если при любых целых $a \geq 1$ уложить хотя бы $K$ плиток нельзя, нужно вывести 0.

## Формат ввода
Во входных данных заданы три целых числа $N, M$ и $K$:
- $1 \leq N, M \leq 10^9$ - размеры стены в сантиметрах,
- $1 \leq K \leq 10^{12}$ - требуемое минимальное количество плиток.

## Формат вывода
Выведите одно целое число - максимальное возможное значение стороны квадратной плитки $a$, при котором на стене можно разместить не менее $K$ плиток.

Если это невозможно, выведите 0.

## Решение

Сделаем бинпоиск по ответу. Если плитка квадратная со стороной $a$, то максимум, сколько мы сможем уложить это $(N//a)*(M//a)$ 

````python
n, m, k = map(int, input().split())

l, r = 1, 10**20  # да, в тупую поставили оч большую границу, но проходит же :)
ans = 0

def can(mid):
    return (n // mid) * (m // mid) >= k

while l + 1 < r:
    mid = (l + r) // 2
    
    if can(mid):
        l = mid
    else:
        r = mid

if n * m < k:
    print(0)
else:
    print(l)
````

# Задача F. Кого больше, мальчиков или девочек?

В Яндекс Лицей на собеседование пришло $N$ школьников, среди которых есть как мальчики, так и девочки.

Преподаватель рассадил их в один ряд за компьютеры.

Для удобства будем считать, что каждый школьник кодируется одним символом:
- символ `а` соответствует мальчику;
- символ `b` соответствует девочке.

К преподавателю подошёл методист и попросил ответить на серию вопросов вида: если взять только школьников с позиции $l$ по позицию $r$ (включительно), кого на этом отрезке больше - мальчиков или девочек, или их количество одинаково?

По заданному ряду школьников и списку запросов нужно для каждого запроса ответить, каких школьников на указанном отрезке больше.

## Формат ввода
В первой строке задано целое число $N\left(1 \leq N \leq 10^6\right)$ - количество школьников в ряду.

Во второй строке записана строка длины $N$, состоящая из символов а и b. Символ `a` соответствует мальчику, символ `b` соответствует девочке.

В третьей строке задано целое число $Q(1 \leq Q \leq 100000)$ - количество вопросов.

В каждой из следующих $Q$ строк заданы два целых числа $l$ и $r(1 \leq l \leq r \leq N)$ - границы отрезка, для которого нужно ответить на вопрос.

## Формат вывода
Для каждого из $Q$ запросов выведите в отдельной строке одно слово:
- `boys`, если на отрезке с $l$ по $r$ строго больше мальчиков, чем девочек;
- `girls`, если на отрезке с $l$ по $r$ строго больше девочек, чем мальчиков;
- `equal`, если количество мальчиков и девочек на этом отрезке одинаково.

## Решение

````python
n = int(input())
s = input().strip()
q = int(input())
````

Если вам в голову пришла корневая декомпозиция, то не надо... Префиксные суммы тут будут проще, хотя бы по написанию кода

Заведем массив префиксных сумм `p` и будем заполнять его во время прохода по строке. Пусть мальчики будут обозначены `1`, а девочки — `-1`. Тогда, если на $i$-й позиции сидит мальчик, то мы делаем `p[i + 1] = p[i] + 1`, а если девочка, то `p[i + 1] = p[i] - 1`

````python
pref = [0] * (n + 1)
for i in range(n):
    if s[i] == 'a':
        pref[i + 1] = pref[i] + 1
    else:
        pref[i + 1] = pref[i] - 1
````

Теперь, когда нас попросят сказать кого больше на отрезке $[l;r]$, мы возьмем `p[r] - p[l - 1]` и если это значение больше 0, значит мальчиков больше, если меньше — девочек, а если равно 0 — поровну

````python
for _ in range(q):
    l, r = map(int, input().split())
    val = pref[r] - pref[l - 1]
    if val > 0:
        print("boys")
    elif val < 0:
        print("girls")
    else:
        print("equal")
````

# Задача G. Рюкзак, максимум стоимости + минимум предметов

Рассмотрим классическую задачу о рюкзаке.

У студента НИУ ВШЭ есть рюкзак, который выдерживает суммарный вес не более $W$. Также есть $N$ предметов. Для каждого предмета известны:
- его вес $w_i$;
- его стоимость $c_i$.

Студент хочет выбрать некоторое подмножество предметов и положить их в рюкзак так, чтобы суммарный вес выбранных предметов не превосходил $W$.

В первую очередь его интересует максимальная возможная суммарная стоимость выбранных предметов. Однако среди всех способов достичь максимальной суммарной стоимости студент хочет выбрать такой, в котором количество предметов минимально.

Допускается, что студент вообще не будет брать никакие предметы. В этом случае суммарная стоимость равна $0$, а количество предметов — $0$.

## Формат ввода
В первой строке заданы два целых числа $N$ и $W$ - количество предметов и максимальная допустимая грузоподъёмность рюкзака $(1 \leq N \leq 100,1 \leq W \leq 1000)$.

Во второй строке записаны $N$ целых чисел $w_i$ - веса предметов $\left(1 \leq w_i \leq 1000\right)$.

В третьей строке записаны $N$ целых чисел $c_i$ - стоимости предметов $\left(1 \leq c_i \leq 1000\right)$.

## Формат вывода
Выведите два целых числа:
- максимальную возможную суммарную стоимость предметов, которые можно положить в рюкзак;
- минимальное количество предметов среди всех наборов, дающих такую максимальную стоимость.

Если ни один предмет не помещается в рюкзак, корректным ответом будет `0 0` (пустой набор предметов: суммарная стоимость $0$, количество предметов $0$).

## Решение

````python
n, w = map(int, input().split())
weights = list(map(int, input().split()))
costs = list(map(int, input().split()))
````

Как и сказано в условии, это классическая задача о рюкзаке. Однако теперь в `dp[i]` мы будем хранить два значения:
- суммарную стоимость
- количество взятых предметов

Если вес `i` недостижим, то стоимость выставим равную `-1`

````python
inf = 10**9
dp = [(-1, inf)] * (w + 1)
dp[0] = (0, 0)  # пустой рюкзак
````

Теперь будем обрабатывать предметы. Для каждого $(w_i,c_i)$ смотрим достижим ли вес $w-w_i$. Если нет, то переходим к следующему предмету. 

Если да, то пробуем положить этот предмет в рюкзак — смотрим на предыдущее состояние рюкзака (без предмета весом $w_i$) `dp[cur_w - wi]`. 

Далее сравним с текущим состоянием `dp[w]`
- если стоимость больше, то это хорошо
- если стоимости равны берем то состояние, в котором меньшее число предметов

Таким образом, среди всех способов с одинаковой стоимостью всегда сохраняем вариант с минимальным количеством предметов

````python
for i in range(n):
    wi = weights[i]
    ci = costs[i]
    for cur_w in range(w, wi - 1, -1):
        prev_val, prev_cnt = dp[cur_w - wi]
        if prev_val == -1:
            continue

        new_val = prev_val + ci
        new_cnt = prev_cnt + 1
        cur_val, cur_cnt = dp[cur_w]

        if new_val > cur_val or (new_val == cur_val and new_cnt < cur_cnt):
            dp[cur_w] = (new_val, new_cnt)
````

Теперь остается среди всего массива `dp` определить максимальное значение веса и минимальное количество предметов. В первую очередь нужно проверять максимальность веса, а потом минимальность количества предметов (в случае равенства весов)

````python
best_val = 0
best_cnt = 0
for val, cnt in dp:
    if val > best_val or (val == best_val and cnt < best_cnt):
        best_val = val
        best_cnt = cnt

print(best_val, best_cnt)
````

# Задача H. НОД на подотрезках

Реализуйте структуру данных для эффективного вычисления НОД нескольких подряд идущих элементов массива.

## Формат ввода
В первой строке вводится одно натуральное число $N(1 \leq N \leq 100000)$ - количество чисел в массиве.

Во второй строке вводятся $N$ чисел от $1$ до $100000$ - элементы массива.

В третьей строке вводится одно натуральное число $K(1 \leq K \leq 30000)$ - количество запросов на вычисление НОД.

В следующих $K$ строках вводится по два числа - номера левого и правого элементов отрезка массива (считается, что элементы массива нумеруются с единицы).

## Формат вывода
Для каждого запроса выведите НОД всех чисел соответствующего участка массива. Числа выводите в одну строку через пробел.

## Решение

Вот тут точно нужна корневая декомпозиция

````python
from math import gcd

n = int(input())
a = list(map(int, input().split()))

size = int(n ** 0.5) + 1
blocks = [0] * ((n + size - 1) // size)

for i in range(n):
    blocks[i // size] = gcd(blocks[i // size], a[i])
````

Поиграем в крутых программистов и напишем отдельную функцию для нахождения НОД на отрезке

Возьмем изначально `res=a[l]`. Мы будем обновлять результат по мере прохода по всему отрезку $[l;r]$

Если границы запроса находятся в одном блоке, то мы не можем взять НОД всего блока и придется посчитать его для всех элементов внутри отрезка

Если же границы в разных блоках, то мы берем НОД блоков, которые находятся внутри краев запроса, а хвосты обрабатываем поэлементно

Отдельно стоит обработать случаи, когда левая граница стоит в начале блкоа:
- если это так, то мы можем использовать НОД всего блока - `blocks[l // size]`, а после сдвинуть `l` на размер блока
- если нет, то придется считать НОд поэлементно, сдвигая левую границу

````python
def query(l, r):
    res = a[l]
    while l <= r:
        if (l // size) != (r // size):
            if l % size:
                res = gcd(a[l], res)
                l += 1
            else:
                res = gcd(blocks[l // size], res)
                l += size
        else:
            res = gcd(a[l], res)
            l += 1
    return res
````

Остается только написать цикл для ввода/вывода

````python
k = int(input())
for _ in range(k):
    l, r = map(int, input().split())
    print(query(l-1, r-1), end=' ')
````
