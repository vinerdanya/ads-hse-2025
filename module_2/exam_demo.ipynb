# Задача A. Делители - количество пар

Дано натуральное число $n\leqslant 10^8$. Подсчитайте количество таких пар чисел $(a,b)$, что:
- $a$ и $b$ - делители $n$
- $a<b$
- $a$ и $b$ - взаимно простые
- $a\cdot b\leqslant n$

```python
n = int(input())
```

По условию, числа $a$ и $b$ - это делители $n$, значит их надо найти

```python
divs = []
for i in range(1, int(n ** 0.5) + 1):
    if n % i == 0:
        divs.append(i)
        divs.append(n // i)
```

Теперь отсортируем `divs`, чтобы выполнить условие, что $a<b$

```python
divs.sort()
```

И дальше рассматриваем перебирать пары $(a, b)$

Так как $j\in \{i+1,\ \text{len(divs)}\}$, это гарантирует, что $a<b$

Также, нужно будет проверить условие, что $a$ и $b$ взаимно простые (то есть `gcd(a,b)=1`) и $a\cdot b\leqslant n$

```python
from math import gcd

cnt = 0
for i in range(len(divs)):
    a = divs[i]
    for j in range(i + 1, len(divs)):
        b = divs[j]
        if gcd(a, b) != 1:
            continue
        if a * b > n:
            break
        cnt += 1

print(cnt)
```

# Задача B. Полная гистограмма при сортировке

## Условие
Дан массив целых чисел $a_1,a_2,\ldots,a_N$, каждый элемент которого лежит в отрезке $[-100;100]$

Требуется для каждого числа $x$ от $-100$ до $100$ посчитать, сколько раз значение $x$ встречается в массиве.

## Формат ввода
В первой строке задано одно целое число $N$ — количество элементов массива $(1\leqslant N\leqslant 10^6)$. Во второй строке заданы $N$ целых чисел $a_i$ $(-100\leqslant a_i\leqslant 100)$

## Формат вывода
Выведите $201$ число — для каждого $x$ от $-100$ до $100$ включительно количество вхождений этого значения в массиве.

Числа необходимо выводить в одной строке, разделяя пробелами. Первое число должно соответствовать значению $x=−100$, второе — значению $x=−99$, и так далее, последнее ($201$-е) число — значению $x=100$

Заведем массив `a` из 201 элемента. Каждый элемент будет отвечать за определенное число на отрезке $[-100;100]$

Тогда, `a[0]` отвечает за число $-100$, `a[1]` — за число $-99$ и т.д. Значит, `a[i + 100]` отвечает за число $i\in[-100;100]$

```python
a = [0] * 201
n = int(input())
b = list(map(int, input().split()))

for i in range(n):
	a[b[i] + 100] += 1

print(*a)
```
